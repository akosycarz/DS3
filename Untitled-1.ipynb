```python

# Create a dictionary to map sports to unique markers
sports_markers = {
    'Sport1': 'o',  # Use 'o' for circles
    'Sport2': 's',  # Use 's' for squares
    'Sport3': '^',  # Use '^' for triangles
    'Sport4': 'D',  # Use 'D' for diamonds
    # Add more sports and corresponding markers as needed
}

# Scatterplot of Age vs. Scores with different markers for sports
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='ind-university_grade', hue='sport', style='sport', palette='Set1')
plt.title('Scatterplot of Age vs. University Grade by Gender and Sport')
plt.xlabel('Age')
plt.ylabel('University Grade')
plt.legend(title='Gender')
plt.show()


```

```python
# Histogram of Age
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Violin Plot of University Grade by Gender
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='gender', y='ind-university_grade', palette='Pastel1')
plt.title('Violin Plot of University Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('University Grade')
plt.show()
```

```python
# Calculate the hiring probability for each combination of age, degree, and gender
hiring_probabilities_age_degree_gender = df.groupby(['age', 'ind-degree', 'gender'])['decision'].mean().reset_index()

# Calculate the hiring probability for each combination of age and programming experience
hiring_probabilities_age_programming = df.groupby(['age', 'ind-programming_exp'])['decision'].mean().unstack()

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Plot the first bar chart (Age, Degree, Gender)
sns.barplot(data=hiring_probabilities_age_degree_gender, x='age', y='decision', hue='gender', ci=None, ax=axes[0])
axes[0].set_title('Probability of Being Hired by Age, Degree, and Gender')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Probability of Being Hired')
axes[0].legend(title='Gender')
axes[0].tick_params(axis='x', rotation=45)

#second bar chart (Age, Programming Experience)
hiring_probabilities_age_programming.plot(kind='bar', ax=axes[1], stacked=True)
axes[1].set_title('Probability of Being Hired by Age and Programming Experience')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Probability of Being Hired')
axes[1].legend(title='Programming Experience')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()




```
